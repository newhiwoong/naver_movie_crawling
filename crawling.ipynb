## 크롤링 + CSV파일로 저장 하는 함수

```python
def FindMovieActor(num):
    global kr, other
    url = "https://movie.naver.com/movie/bi/mi/detail.nhn?code="
    url = url + str(num)
    try:
        html = urlopen(url)
    except urllib.error.HTTPError as e:
        print(e)
        print("=============error %d ============="%num)
    else:
        try:
            bsObj = BeautifulSoup(html.read(), "html.parser")
        except urllib.exceptions.SSLError as e:
            print(e)
            print("=============error %d ============="%num)
        else:
            #print("============%d===================" % num)
            check_KR = bsObj.find("a",{"href":"/movie/sdb/browsing/bmovie.nhn?nation=KR"})
            year = bsObj.find("strong",{"class":"h_movie2"})
            if(not year):
                return
            year = year.get_text()[-4:]
            try:
                int(year)
            except ValueError:
                return
            actors = bsObj.findAll("a",{"class":"k_name"})
            summarys = bsObj.findAll("a",{"href":re.compile("/movie/sdb/browsing/bmovie.nhn\?genre=*")})
            #print(check_KR)
            if(check_KR and int(year) >= 2010 and len(actors) >= 3 and summarys):
                csvRow = []
                findname = "./basic.nhn?code="+str(num)
                name = bsObj.find("a",{"href":findname}).get_text()
                director = bsObj.find("a",{"href":re.compile("/movie/bi/pi/basic\.nhn\?code=*")}).get_text()
                ratings = bsObj.findAll("em",{"class":"p_part"})
                print("============%d===================" % num)
                #print(name)
                csvRow.append(name)
                csvRow.extend([year,director,str(len(summarys)/2),str(len(actors)-1)])
                
                for i in range(int(len(summarys)/2)):
                    csvRow.append(summarys[i].get_text())
                    
                for actor,rating in zip(actors,ratings):
                    #print(actor.get_text()," : ",rating.get_text())
                    csvRow.append(actor.get_text())
                    csvRow.append(rating.get_text())   
                #print(csvRow)
                try:
                    writer.writerow(csvRow)
                except UnicodeEncodeError:
                    print("============error white%d===================" % num)
                else:
                    kr+=1
            else:
                #print("no")
                other+=1

        
```
