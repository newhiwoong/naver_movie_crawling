## 저장한 CSV 불려오고 list로 만들기

```python
import csv
f = open("files/movie.csv", 'r',encoding='euc_kr')
reader = csv.reader(f)
MovieToActor = []
for line in reader:
    MovieToActor.append(line)
f.close()
```

```python
print(len(MovieToActor),len(MovieToActor[0]))

#print(MovieToActor)
```

```python
MovieToActor[37][5+int(MovieToActor[2][3]):]
```

```python
Actors = []

j,k = 0,0

for i in range(1,len(MovieToActor)):
    #print(i)
    j= 5+int(MovieToActor[i][3])
    while(1):
        if j >= len(MovieToActor[i]):
            break
        if MovieToActor[i][j] == '':
                break
        k=5+int(MovieToActor[i][3])
        while(1):
            if k >= len(MovieToActor[i]):
                break
            if MovieToActor[i][k] == '':
                break
            if not j==k:
                Actors.append([MovieToActor[i][j],MovieToActor[i][k]])
            k+=2
            
        j+=2
        if j >= len(MovieToActor[i]):
            break

        
```

```python
Actors
```

```python
csvFile = open("files/ActorPairs2.csv", 'w',encoding='euc_kr')
writer = csv.writer(csvFile)
for csvRow in Actors:
    writer.writerow(csvRow)
csvFile.close()
```
